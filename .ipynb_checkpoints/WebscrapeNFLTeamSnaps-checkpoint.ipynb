```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

page = 0
login_url = "https://stathead.com/users/login.cgi"
payload = {
    '_username':'tim@fantasysidelines.com',
    '_password':'tC322011'
}
url = "https://stathead.com/football/tgl_finder.cgi?\
request=1&match=game&order_by_asc=0&order_by=plays_offense&year_min=2017&year_max=2019&\
game_type=R&game_num_min=0&game_num_max=99&week_num_min=0&week_num_max=99&temperature_gtlt=lt&offset="
```

```python
#1500
session = requests.Session()
s = session.post(login_url, data=payload)
while page < 1600:
    page1 = str(page)
    website = session.get(url+page1).text
    soup = BeautifulSoup(website, 'lxml')

    table = soup.find('table', attrs={'class': 'sortable', 'id': 'results'})

    table_headers = [header.text for header in table.find('thead').find_all('th')]
    table_rows = table.find_all('tr')

    final_data = []
    for tr in table_rows:
        td = tr.find_all('td')
        row = [tr.text for tr in td]
        final_data.append(row)
    df = pd.DataFrame(final_data, columns=table_headers[3:])
    df.to_csv('nflteamsnaps.csv',mode='a',header=False)
    page += 100
    print(page)
```
