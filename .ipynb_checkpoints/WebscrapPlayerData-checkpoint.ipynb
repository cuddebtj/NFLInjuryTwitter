```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import os
from dotenv import load_dotenv
load_dotenv()

page = 0
login_url = "https://stathead.com/users/login.cgi"
user_name = os.environ.get('statheadusername')
password = os.environ.get('statheadpassword')
payload = {
    'username': user_name,
    'password': password
}
url = "https://stathead.com/football/pgl_finder.cgi?request=1&game_num_max=99&week_num_max=99&order_by=all_td&season_start=1\
&qb_gwd=0&order_by_asc=0&qb_comeback=0&week_num_min=0&game_num_min=0&year_min=2017&match=game&year_max=2019&season_end=-1\
&age_min=0&game_type=R&age_max=99&positions[]=qb&positions[]=rb&positions[]=wr&positions[]=te&positions[]=e&positions[]=t\
&positions[]=g&positions[]=c&positions[]=ol&positions[]=dt&positions[]=de&positions[]=dl&positions[]=ilb&positions[]=olb\
&positions[]=lb&positions[]=cb&positions[]=s&positions[]=db&positions[]=k&positions[]=p&cstat[1]=two_pt_md&ccomp[1]=gt\
&cval[1]=0&cstat[2]=rush_att&ccomp[2]=gt&cval[2]=0&cstat[3]=pass_att&ccomp[3]=gt&cval[3]=0&cstat[4]=kick_ret_td&ccomp[4]=gt\
&cval[4]=0&cstat[5]=fumbles_rec_td&ccomp[5]=gt&cval[5]=0&cstat[6]=targets&ccomp[6]=gt&cval[6]=0&cstat[7]=touches&ccomp[7]=gt\
&cval[7]=0&cstat[8]=punt_ret_yds_per_ret&ccomp[8]=gt&cval[8]=0&offset="
```

```python
with requests.Session() as session:
    s = session.post(login_url, data=payload)
    while page < 61500:
        page1 = str(page)
        website = session.get(url+page1).text
        soup = BeautifulSoup(website, 'html')

        table = soup.find('table', attrs={'class': 'sortable', 'id': 'results'})

        table_headers = [header.text for header in table.find('thead').find_all('th')]
        table_rows = table.find_all('tr')

        final_data = []
        for tr in table_rows:
            td = tr.find_all('td')
            row = [tr.text for tr in td]
            final_data.append(row)
        df = pd.DataFrame(final_data, columns=table_headers[8:])
        df.to_csv('nfldata.csv',mode='a',header=False)
        page += 100
        print(page)
```
