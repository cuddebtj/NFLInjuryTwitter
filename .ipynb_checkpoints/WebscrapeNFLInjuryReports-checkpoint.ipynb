```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

#list of the NFL teams for url purposes
teams = ['crd', 'atl', 'rav', 'buf', 'car', 'chi', 'cin', 'cle', 'dal', 'den', 'det', 'gnb','htx','clt','jax','kan',
         'sdg','ram','mia','min','nor','nwe','nyg','nyj','rai','phi','pit','sea','sfo','tam','oti','was']

#list of years to pull data for url purposes
years= ['2017','2018', '2019']

#starting points to iterate through
team = 1
year = 0
num = list(range(30000))
```

```python
while team < 2:
    #url for web scraping
    url = f'https://www.pro-football-reference.com/teams/{teams[team]}/{years[year]}_injuries.htm'

    #opening website
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'lxml')

    #finding table
    table = soup.find('table', attrs={'class': 'sortable', 'id': 'team_injuries'})
    
    #creating lists to hold scrape data
    week = []
    injury = []
    player0 = []
    
    #scraping the data
    #having issue with playoff games being on injury report
    for attr in table.find_all('td'):
        week.append(attr.get('data-stat'))
        injury.append(attr.get('data-tip'))

    for attr in table.find('tbody').find_all('th'):
        player0.append(attr.text)
    
    #organizing data for dataframe
    player1 = player0.copy()
    player2 = player0.copy()
    player3 = player0.copy()
    player4 = player0.copy()
    player5 = player0.copy()
    player6 = player0.copy()
    player7 = player0.copy()
    player8 = player0.copy()
    player9 = player0.copy()
    player10 = player0.copy()
    player11 = player0.copy()
    player12 = player0.copy()
    player13 = player0.copy()
    player14 = player0.copy()
    player15 = player0.copy()

    player = []
    for a, b, c, d, e, f, g, h, i, j, k, l, m, n, o, p in zip(player0, player1, player2, player3, player4, player5, player6, player7, player8, player9, player10, player11, player12, player13, player14, player15):
        player.append(a)
        player.append(b)
        player.append(c)
        player.append(d)
        player.append(e)
        player.append(f)
        player.append(g)
        player.append(h)
        player.append(i)
        player.append(j)
        player.append(k)
        player.append(l)
        player.append(m)
        player.append(n)
        player.append(o)
        player.append(p)

    #creating dict for dataframe and creating csv to save dict
    data = {num[i]: [player[i],teams[team], years[year], week[i], injury[i]] for i in range(0, len(player))}
    df = pd.DataFrame(data)
    df = df.T
    df.to_csv('nflinjuryreport.csv',mode='a',header=False)

    #advancing through url
    if year < 2:
        year += 1
        
    else:
        year = 0
        team += 1
    
    #delete lists for re-recreation
    del week
    del injury
    del player0
    i = 0

    # update of location of web scrape
    print(teams[team])
    print(years[year])
```
