```python
# https://github.com/geduldig/TwitterAPI
from TwitterAPI import TwitterAPI
import json
import pandas as pd
import os
from dotenv import load_dotenv
load_dotenv()

consumer_api_key = os.environ.get('Twitter_API_Key')
consumer_api_secret = os.environ.get('Twitter_API_Secret_Key')
access_token_key = os.environ.get('Twitter_Access_Token')
access_token_secret = os.environ.get('Twitter_Access_Token_secret')
```

```python
api = TwitterAPI(consumer_key=consumer_api_key,
                 consumer_secret=consumer_api_secret,
                 access_token_key=access_token_key,
                 access_token_secret=access_token_secret)
PRODUCT = 'fullarchive'
LABEL = 'injuries' # This is specific to your application
# i.e. whatever label you set for your Dev environment, and is case sensitive
# The following should be uncommented for the first web request.
# This code was run in a Python notebook, in separate cells.
# It might need to be restructured for other uses.
list_tweets = []
next = None
web_request_count = 0
while next is not None:
    r = api.request('tweets/search/%s/:%s' % (PRODUCT, LABEL),        
           {'query':'Christian McCaffrey'
           }) 
    print('r.status_code: ', r.status_code)
    next = r.json()['next']
    print('web_request_count: ', web_request_count)
    print('r.get_quota: ', r.get_quota)
    web_request_count += 1
    results = r.json()['results']
    for tweet in results:
        twitter_handle = tweet['screen_name']['screen_name']
        twitter_followers = tweet['followers_count']['followers_count']
        tweet_date = tweet['created_at']
        tweet_text = tweet['text']
        if 'extended_tweet' in tweet.keys():
            tweet_text = tweet['extended_tweet']['full_text']
            tweet_row = {'user':twitter_handle,
                         'followers': twitter_followers,
                         'date':tweet_date,
                         'full_text':tweet_text}
            list_tweets.append(tweet_row)
    df = pd.DataFrame(list_tweets)

    df.to_csv('test_tweets.csv', mode='a',index=False)
    time.sleep(2.1) # wait i.e. only 30 requests per minute allowed
```
